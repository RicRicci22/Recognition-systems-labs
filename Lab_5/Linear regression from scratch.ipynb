```python
import numpy as np 
import random
import matplotlib.pyplot as plt
```

```python
random.seed(45)
np.random.seed(45)
```

```python
N = 100
X = np.random.rand(N,1)
y = np.zeros((N,1))

y_values = [i+random.random() for i in X[:,0]]

y[:,0] = y_values

plt.scatter(X[:,0],y)
plt.show()
```

```python
X_tot = np.append(np.ones((N,1)),X,axis=1)
```

```python
first_term = np.linalg.inv(np.dot(X_tot.T,X_tot))
print(first_term.shape)
```

```python
second_term = np.dot(X_tot.T,y)
print(second_term.shape)
```

```python
optimal_weights = np.dot(first_term,second_term)
print(optimal_weights.shape)
```

```python
print(optimal_weights)
```

```python
X_temp=np.ones((1000,2))
x_values = [i/1000 for i in range(0,1000)]
X_temp[:,1] = x_values

y_hat = np.dot(X_temp,optimal_weights)
print(y_hat.shape)
```

```python
plt.scatter(X,y)
plt.plot(X_temp[:,1],y_hat,color='red')
plt.show()
```
