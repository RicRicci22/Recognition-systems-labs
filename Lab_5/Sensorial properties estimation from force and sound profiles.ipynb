```python
from scipy import interpolate

def interpolate_spline(in_x,in_y,x_bounds,output_length):
  s = interpolate.InterpolatedUnivariateSpline(in_x,in_y)
  x_fit = np.linspace(x_bounds[0],x_bounds[1],output_length)
  y_fit = s(x_fit)
  return (x_fit,y_fit)
```

```python
N_points = 200

for sample in obj['Texture_measures'].keys():
  force = obj['Texture_measures'][sample]['Force']
  strain = obj['Texture_measures'][sample]['Strain']

  interpolated_strain, interpolated_force = interpolate_spline(strain,force,(0,90),N_points)

  plt.figure(figsize=(25,12))
  plt.subplot(1,2,1)
  plt.plot(strain,force,label='Original')
  plt.legend()
  plt.subplot(1,2,2)
  plt.plot(interpolated_strain,interpolated_force,label='Interpolated')
  plt.legend()
  plt.show()
  break

print(len(force))
print(len(interpolated_force))
```

```python
plt.figure(figsize=(25,12))
plt.plot(strain,force,label='Original')
plt.plot(interpolated_strain,interpolated_force,label='Interpolated')
plt.legend()
plt.show()
```

```python
processed_force = np.zeros((len(os.listdir(path_raw_data)),200))
processed_sound = np.zeros((len(os.listdir(path_raw_data)),200))

# Targets 
hardness = np.zeros((len(os.listdir(path_raw_data)),1))
graininess = np.zeros((len(os.listdir(path_raw_data)),1))

i = 0 
for filename in os.listdir(path_raw_data):
  with open(os.path.join(path_raw_data,filename),'rb') as data:
    obj = pickle.load(data)
  
  temp_force = np.zeros((len(obj['Texture_measures'].keys()),200))
  temp_sound = np.zeros((len(obj['Texture_measures'].keys()),200))

  j = 0 
  for key in obj['Texture_measures'].keys():
    strain = obj['Texture_measures'][key]['Strain']
    force = obj['Texture_measures'][key]['Force']
    sound = obj['Texture_measures'][key]['Sound']

    interpolated_strain, interpolated_force = interpolate_spline(strain,force,(0,90),N_points)
    interpolated_strain, interpolated_sound = interpolate_spline(strain,sound,(0,90),N_points)

    temp_force[j,:] = interpolated_force
    temp_sound[j,:] = interpolated_sound
    j+=1
  
  processed_force[i,:] = np.mean(temp_force,axis=0)
  processed_sound[i,:] = np.mean(temp_sound,axis=0)

  with open(os.path.join(path_sensory_data,filename),'rb') as data:
    obj = pickle.load(data)

  hardness[i,0] = obj['Hardness']
  graininess[i,0] = obj['Graininess']

  i+=1
```

```python
plt.figure(figsize=(25,12))
plt.subplot(2,1,1)
plt.plot(np.linspace(0,90,200),processed_force[0,:])
plt.subplot(2,1,2)
plt.plot(np.linspace(0,90,200),processed_sound[0,:])
plt.show()
```

```python
import random 

train_portion = 0.7
random.seed(15)

# Shuffle the dataset 
indices = [i for i in range(processed_force.shape[0])]
print(indices)
random.shuffle(indices)
print(indices)
```

```python
train_indices = indices[0:int(processed_force.shape[0]*train_portion)]
test_indices = indices[int(processed_force.shape[0]*train_portion):]
```

```python
# Features 
train_force = processed_force[train_indices,:]
train_sound = processed_sound[train_indices,:]
test_force = processed_force[test_indices,:]
test_sound = processed_sound[test_indices,:]
print(train_force.shape)
print(test_force.shape)
```

```python
# Targets
train_hardness = hardness[train_indices,:]
train_graininess = graininess[train_indices,:]
test_hardness = hardness[test_indices,:]
test_graininess = graininess[test_indices,:]
print(train_hardness.shape)
print(test_hardness.shape)
```

```python
# Normalize the data
min_force = np.min(np.min(train_force))
max_force = np.max(np.max(train_force))

min_sound = np.min(np.min(train_sound))
max_sound = np.max(np.max(train_sound))
```

```python
# Transform
train_force = (train_force-min_force)/(max_force-min_force)
train_sound = (train_sound-min_sound)/(max_sound-min_sound)
```

```python
test_force = (test_force-min_force)/(max_force-min_force)
test_sound = (test_sound-min_sound)/(max_sound-min_sound)
```

```python
train_hardness = train_hardness/100
test_hardness = test_hardness/100
train_graininess = train_graininess/100
test_graininess = test_graininess/100
```

```python
plt.figure(figsize=(25,12))
plt.subplot(2,1,1)
plt.plot(np.linspace(0,90,200),train_force[0,:])
plt.subplot(2,1,2)
plt.plot(np.linspace(0,90,200),train_sound[0,:])
plt.show()
```

```python
# FEATURE EXTRACTION 
def get_max(in_matrix):
  maximums = np.max(in_matrix,axis=1)
  return maximums
```

```python
force_max = get_max(train_force)
```

```python
X_train = np.zeros((train_force.shape[0],1))
X_train[:,0] = force_max
```

```python
force_max = get_max(test_force)
X_test = np.zeros((test_force.shape[0],1))
X_test[:,0] = force_max
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from scipy.stats import pearsonr

plt.figure(figsize=(20,8))
plt.subplot(1,2,1)
plt.scatter(X_train[:,0],train_hardness)
plt.show()
```

```python
pearson_coeff,_ = pearsonr(X_train[:,0],train_hardness)
print(pearson_coeff)
```

```python
plt.figure(figsize=(20,8))
plt.subplot(1,2,1)
plt.scatter(X_train[:,0],train_graininess)
plt.show()
```

```python
pearson_coeff,_ = pearsonr(X_train[:,0],train_graininess)
print(pearson_coeff)
```

```python
# REGRESSION
from sklearn.linear_model import LinearRegression
# HARDNESS
linear_regressor = LinearRegression(fit_intercept=True)
linear_regressor.fit(X_train,train_hardness)
predictions_hardness_linear = linear_regressor.predict(X_test)
# GRAININESS
linear_regressor = LinearRegression(fit_intercept=True)
linear_regressor.fit(X_train,train_graininess)
predictions_graininess_linear = linear_regressor.predict(X_test)
```

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.plot([0,1],[0,1])
plt.scatter(test_hardness,predictions_hardness_linear)
plt.subplot(1,2,2)
plt.scatter(test_graininess,predictions_graininess_linear)
plt.plot([0,1],[0,1])
plt.show()
```
